```python
import pandas as pd
import matplotlib.pyplot as plt

try:
    # Read the data from the file generated by the C++ program
    data = pd.read_csv('expense_data.txt', header=None, names=['Category', 'Amount'])

    # Create the pie chart
    plt.figure(figsize=(8, 8))
    plt.pie(data['Amount'], labels=data['Category'], autopct='%1.1f%%', startangle=90, colors=plt.cm.Paired.colors)
    plt.title('Expense Breakdown by Category')
    plt.axis('equal')  # Equal aspect ratio ensures the pie is circular.

    # Save the pie chart as an image file
    plt.savefig('expense_piechart.png')
    plt.show() # Display the chart window

    print("Pie chart saved as 'expense_piechart.png'")

except FileNotFoundError:
    print("Error: 'expense_data.txt' not found. Please run the C++ program first to generate the data.")
except Exception as e:
    print(f"An error occurred: {e}")
```
